# *Reminder of Itô’s lemma*
## The part that is Brownian Specific
Kiyoshi Itô (1940s) was building a new **integration theory** for stochastic processes.  

- Ordinary calculus: $\int f(t)\,dt$  
- Stochastic calculus: $\int f(t, W_t)\, dW_t$, where $dW_t^{\mathbb P} \equiv W_{t+dt}^{\mathbb P} - W_t^{\mathbb P}, \quad
dW_t^{\mathbb P} \sim \mathcal N(0,dt).$

But Brownian motion $W_t$ is nowhere differentiable, so classical integration does not work. Itô’s main achievement was to define a new type of integral.

## 2. The “main theorem”
The central result is the **existence and properties of the Itô integral**.  

For a square-integrable process $f$, the integral
\begin{equation}
\int_0^t f(s, W_s)\, dW_s
\end{equation} 
is well-defined, and it satisfies:

- **Martingale property**  
  The stochastic integral is a martingale.  

    -- 1. $\mathbb{E}[|X_t|] < \infty$ for all $t$.
    -- 2.	For all $s<t$,  $\mathbb{E}[X_t \mid \mathcal{F}_s] = X_s.$

- **Itô isometry**  -- For Brownian Motion
  \begin{equation}
  \mathbb{E}\!\left[\left(\int_0^t f(s, W_s)\, dW_s\right)^2\right]
  = \mathbb{E}\!\left[\int_0^t f(s, W_s)^2\, ds\right].
  \end{equation}
  
  \begin{equation}
  \int_a^b f(t)\, dW_t
    = \lim_{\|\Pi\|\to 0} \sum_{i=0}^{n-1} f(t_i)\,\big(W_{t_{i+1}}-W_{t_i}\big),
  \end{equation}
  The integral is the sum of all the $f(t)$, weighted by the incremental (infinitesimally small) of the random part, similar to the chain rule where $dW_t = dW_t/dt\times dt$ not the possible $W_t$ at time $t$, but rather $W_t = W(t)$. 
   
This theorem is the foundation of stochastic calculus.


## 3. Where Itô’s lemma fits
Once the integral is defined, consider an SDE
\begin{equation}
dX_t = a(X_t,t)\,dt + b(X_t,t)\,dW_t.
\end{equation}

If $f=f(X_t,t)$, what dynamics does $f$ satisfy?  

**Itô’s lemma** gives the answer:
\begin{equation}
df = f_t\,dt + f_x\,dX_t + \tfrac{1}{2} f_{xx}\,(dX_t)^2.
\end{equation}

This is the **chain rule** in stochastic calculus, derived from the Itô integral and the quadratic variation property of Brownian motion.

- In real analysis:
  - **Main theorem**: the Lebesgue integral exists with convergence theorems.   
  - **Chain rule**: a lemma built upon it.  

- In stochastic calculus:
  - **Main theorem**: the Itô integral is well-defined with isometry and martingale properties. 

    -- 1. $\mathbb{E}[|X_t|] < \infty$ for all $t$.
    -- 2.	For all $s<t$,  $\mathbb{E}[X_t \mid \mathcal{F}_s] = X_s.$
  - **Itô’s lemma**: the chain rule for SDEs.

\begin{equation}
dX_t = a\,dt + b\,dW_t.
\end{equation}
- The deterministic part $a\,dt$ is order $dt$.
- The random part $b\,dW_t$ is order $\sqrt{dt}$. 
This can be seen in two ways:

***
## Stories in our story
First:
\begin{equation}
\mathbb{E}[|dW_t|] = \sqrt{dt}\,\mathbb{E}[|Z|] = \sqrt{dt}\,\sqrt{\tfrac{2}{\pi}}
\end{equation}

Second,
Because $dW_t/\sqrt{dt} \Rightarrow \mathcal N(0,1)$ (tight, non-degenerate), we write $dW_t = O_p(\sqrt{dt}).$
That is, for any $\varepsilon>0$ there exists $M$ s.t.

\begin{equation}
\Pr\!\left(\frac{|dW_t|}{\sqrt{dt}} > M\right) < \varepsilon
\end{equation}
for all sufficiently small dt. This formalizes “order $\sqrt{dt}$.”

Quick flashback:
\begin{equation}
\begin{aligned}
\text{Deterministic:} & \forall \delta>0, \exists M:\; 0<h<\delta \implies |f(h)| \le M h^\alpha \implies f(h) = \mathcal O(h^\alpha). \\
\text{Stochastic:} & \forall \varepsilon>0, \exists M:\; \Pr\!\left( |X_h| > M h^\alpha \right) < \varepsilon \quad(h \text{ small}) \implies X_h = \mathcal O(h^\alpha). \\
\end{aligned}
\end{equation}
(It is crucial to find the smallest $\alpha$ to correctly to define the infinitesimal behavior)
So the dominant scaling of $dX_t$ is $\sqrt{dt}$, not $dt$.

***

1. Classical Taylor expansion

For a smooth $f(x,t)$:

\begin{equation}
df = f_t\,dt + f_x\,dX_t + \tfrac{1}{2} f_{xx} (dX_t)^2 + o\!\big((dX_t)^2\big).
\end{equation}

In ordinary calculus, if $dX_t \sim dt$, then
- $(dX_t)^2 \sim (dt)^2$, which is negligible compared to $dt$.
- So we drop both $\tfrac{1}{2} f_{xx}(dX_t)^2$ and the remainder $o((dX_t)^2)$.

Substituting the stochastic process (GBM)
Square it:
\begin{equation}
(dX_t)^2 = (a\,dt + b\,dW_t)^2= a^2 (dt)^2 + 2ab\,dt\,dW_t + b^2(dW_t)^2.
\end{equation}
### Apply Itô rules “Itô’s lemma = Taylor expansion truncated at second order”
- $(dt)^2 = 0,$
- $dt\,dW_t = 0,$
- $(dW_t)^2 = dt.$

Itô’s lemma is similar: it looks like a truncation trick, but it’s actually the rigorous foundation that allows us to define stochastic integrals, PDE connections, risk-neutral pricing, etc.

Proof of the third one:

***

Let $\Delta W := W_{t+\Delta t}-W_t \sim \mathcal N(0,\Delta t)$. Then

\begin{equation}
\mathbb{E}\!\left[(\Delta W)^2\right]=\Delta t,\qquad
\mathrm{Var}\!\left((\Delta W)^2\right)
= \mathbb{E}[(\Delta W)^4]-\big(\mathbb{E}(\Delta W)^2\big)^2
= 3(\Delta t)^2-(\Delta t)^2 = 2(\Delta t)^2.
\end{equation}
Hence, with $R_{\Delta t}:=(\Delta W)^2-\Delta t$,

\begin{equation}
\mathbb{E}[R_{\Delta t}]=0,\qquad
\mathrm{SD}(R_{\Delta t})=\sqrt{2}\Delta t.
\end{equation}
So $R_{\Delta t}=O_p(\Delta t)$ (indeed also $O_{L^2}(\Delta t)$).
Conclusion: for a single increment,
\begin{equation}
(\Delta W)^2 = \Delta t + O_p(\Delta t),
\end{equation}
not $O(\Delta t^2)$.

***
Alternatively,

Let $\Pi=\{0=t_0<t_1<\dots<t_n=T\}$ be a partition, and define
\begin{equation}
Q(\Pi)=\sum_{i=0}^{n-1}\big(W_{t_{i+1}}-W_{t_i}\big)^2.
\end{equation}
For Brownian motion,
\begin{equation}
Q(\Pi)\xrightarrow[|\Pi|\to 0]{\text{a.s.}} T,
\end{equation}
Where $\|\Pi\| := \max_i (t_{i+1}-t_i).$ so the quadratic variation is $[W]_t=t$. Heuristically this is the integral identity
\begin{equation}
\sum (dW_t)^2 = \int_0^T dt,
\quad\text{ i.e. }\quad (dW_t)^2=dt.
\end{equation}

# The Lemma Beyond Brownian

Finance quickly saw that real markets are not pure Brownian:
- Heavy tails → models with jumps (Poisson, Lévy).
- Volatility clustering → stochastic volatility (Heston model).
- Long memory → fractional Brownian motion and rough volatility.

In these cases, Itô’s lemma still works, but in generalized forms (with quadratic variation and jump terms).

# Black–Scholes PDE: step-by-step derivation

## 1) Modeling assumptions
- Risky asset $S_t$ follows GBM (Geometric Brownian Motion):
  \begin{equation}
  dS_t = \mu S_t\,dt + \sigma S_t\,dW_t, \quad \sigma>0
  \end{equation}
- Money market account $B_t$ grows at constant risk-free rate $r$:
  \begin{equation}
  dB_t = r B_t\,dt
  \end{equation}
- No arbitrage, continuous trading, no dividends.  
- Derivative value $V=V(S,t)$ is smooth enough ($C^{2,1}$ in $(S,t)$).
- Where $dW_t \sim \mathcal N(0,dt)$


## 2) Apply Itô’s lemma to $V(S_t,t)$ -- the equation for the derivatives -- total value(time value + intrinsic value)

\begin{equation}
dV = \frac{\partial V}{\partial t}\,dt 
   + \frac{\partial V}{\partial S}\,dS_t 
   + \tfrac{1}{2}\frac{\partial^2 V}{\partial S^2}(dS_t)^2
\end{equation}

Since $(dW_t)^2 = dt$ and $(dS_t)^2 = \sigma^2 S_t^2 dt$:

\begin{equation}
dV = \left(\frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} 
      + \tfrac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}\right)dt
   + \sigma S \frac{\partial V}{\partial S}\,dW_t
\end{equation}


## 3) Construct hedged portfolio
Portfolio:
\begin{equation}
\Pi_t = V(S_t,t) - \Delta S_t
\end{equation}
- $V(S_t,t)$ is the value of the option.
- $\Pi_t$ is the hedged portfolio (long the option, short $\Delta$ shares).
- $\Delta$ is the hedge ratio (also called option delta in Greeks)

Self-financing change:
\begin{equation}
d\Pi = dV - \Delta\,dS
\end{equation}

Choose $\Delta = \frac{\partial V}{\partial S}$ so the $dW_t$ term cancels:
\begin{equation}
d\Pi = \left(\frac{\partial V}{\partial t} 
        + \tfrac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}\right)dt
\end{equation}



## 4) No-arbitrage condition
Riskless $\Pi$ must earn risk-free rate:
\begin{equation}
d\Pi = r\Pi\,dt = r\left(V - S\frac{\partial V}{\partial S}\right)dt
\end{equation}

Equating both:
\begin{equation}
\frac{\partial V}{\partial t} + \tfrac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
= r\left(V - S\frac{\partial V}{\partial S}\right)
\end{equation}

## 5) We arrive at Black–Scholes PDE

\begin{equation}
\frac{\partial V}{\partial t}
+ \tfrac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS\frac{\partial V}{\partial S} - rV = 0
\end{equation}



## 6) Terminal & boundary conditions -- European Style
European call ($K,T$):
\begin{equation}
V(S,T) = \max(S-K,0)
\end{equation}

\begin{equation}
V(0,t)=0, \quad V(S,t)\sim S - K e^{-r(T-t)} \ \text{as } S\to\infty
\end{equation}

European put:
\begin{equation}
V(S,T) = \max(K-S,0)
\end{equation}

\begin{equation}
V(0,t)=K e^{-r(T-t)}, \quad V(S,t)\to 0 \ \text{as } S\to\infty
\end{equation}

## 7) Risk-neutral representation
the stock dynamics under the risk-neutral measure $\mathbb{Q}$:
\begin{equation}
dS_t = r S_t\,dt + \sigma S_t\,dW_t^{\mathbb{Q}}
\end{equation}

$dW_t^{\mathbb{Q}} = dW_t^{\mathbb{P}} + \theta\, dt,$ where $\theta = \frac{\mu - r}{\sigma}$. The reason for using $\mathbb Q-$ martingale is because the risk-neutral hypothesis, and we move all the premiums inside the measures/probability density $\mathbb Q$.

$W^{\mathbb P}$ are $W^{\mathbb Q}$ two different process，they are both $\mathcal N(0,t)$ in their measure.
analoge to coin toss:
- $\mathbb P: p(H) = 0.6$ $W_n=\sum_{i}^n (X_i-0.2)$
- $\mathbb Q: p(H) = 0.5$ $W_n=\sum_{i}^n X_i$, $\mathbb E^{\mathbb P}[W_n] \neq 0$, non martingale, but $\mathbb E^{\mathbb Q}[W_n] = 0$, martingale
- ($\mathbb P,\,\mathbb Q$: measure/probability distribution)


By Feynman–Kac:
\begin{equation}
V(S,t) = e^{-r(T-t)} \, \mathbb{E}^{\mathbb{Q}} \big[\Phi(S_T) \mid S_t=S \big]
\end{equation}
- $\Phi$ Payoff function


## 8) Closed-form Black–Scholes formula
1) Risk-neutral distribution of $S_T$
\begin{equation}
\ln S_T \sim \mathcal N\!\Big(\ln S + (r-\tfrac{1}{2}\sigma^2)\tau,\; \sigma^2 \tau\Big),
\end{equation}
equivalently
\begin{equation}
S_T = S \exp\!\Big( (r-\tfrac{1}{2}\sigma^2)\tau + \sigma\sqrt{\tau}\, Z \Big), 
\quad Z \sim \mathcal N(0,1).
\end{equation}

2) Call option value as discounted expectation
\begin{equation}
C = e^{-r\tau}\, \mathbb E^{\mathbb Q}\!\left[ (S_T-K)^+ \right].
\end{equation}

Split into two parts:
\begin{equation}
C = e^{-r\tau}\left( \mathbb E[S_T \mathbf 1_{\{S_T>K\}}] - K\, \mathbb Q(S_T>K) \right).
\end{equation}

3) Compute probabilities
\begin{equation}
\mathbb Q(S_T>K) = \mathbb Q\!\left(Z > \frac{\ln K - \ln S - (r-\tfrac12\sigma^2)\tau}{\sigma\sqrt{\tau}} \right) 
= N(d_2),
\end{equation}
with
\begin{equation}
d_2 = \frac{\ln(S/K) + (r-\tfrac12\sigma^2)\tau}{\sigma\sqrt{\tau}}.
\end{equation}

\begin{equation}
\mathbb E[S_T \mathbf 1_{\{S_T>K\}}] = S e^{r\tau} N(d_1),
\end{equation}
where
\begin{equation}
d_1 = \frac{\ln(S/K) + (r+\tfrac12\sigma^2)\tau}{\sigma\sqrt{\tau}} = d_2 + \sigma\sqrt{\tau}.
\end{equation}

4) Call price
\begin{equation}
C = S N(d_1) - K e^{-r\tau} N(d_2).
\end{equation}

5) Put option via put–call parity
\begin{equation}
P = K e^{-r\tau} N(-d_2) - S N(-d_1).
\end{equation}

## Parameters

```python
S0, K = 100.0, 100.0
r, sigma = 0.05, 0.20
T = 1.0
```

## Black–Scholes closed-form (inline)
Call:

\begin{equation}
C = S_0 N(d_1) - K e^{-rT} N(d_2),\quad
d_1=\frac{\ln(S_0/K)+(r+\tfrac{1}{2}\sigma^2)T}{\sigma\sqrt{T}},\;
d_2=d_1-\sigma\sqrt{T}.
\end{equation}

Put:

\begin{equation}
P = K e^{-rT} N(-d_2) - S_0 N(-d_1).
\end{equation}

```python
true_call = bs_call(S0, K, r, sigma, T)
true_put  = bs_put (S0, K, r, sigma, T)
true_call, true_put
```

## Monte Carlo estimator
Discounted payoff:

\begin{equation}
\hat{C}_{\mathrm{MC}} = e^{-rT}\,\frac{1}{N}\sum_{i=1}^N (S_T^{(i)}-K)^+.
\end{equation}

Standard error and 95% CI:

\begin{equation}
\mathrm{SE}=\frac{s}{\sqrt{N}},\quad
\text{CI}_{95\%}=\hat{C}\pm 1.96\,\mathrm{SE}.
\end{equation}

```python
def mc_euro_call(S0, K, r, sigma, T, n, rng=None):
    ST = sample_terminal_risk_neutral(S0, r, sigma, T, n_paths=n, rng=rng)
    payoff = np.maximum(ST - K, 0.0)
    disc_payoff = np.exp(-r * T) * payoff
    price = disc_payoff.mean()
    se = disc_payoff.std(ddof=1) / math.sqrt(n)
    return price, se

# Quick smoke test
mc_price, mc_se = mc_euro_call(S0, K, r, sigma, T, n=200_000, rng=rng)
mc_price, mc_se, true_call
```

## CI coverage check
Does the 95% CI include the analytic price?

```python
lo, hi = mc_price - 1.96 * mc_se, mc_price + 1.96 * mc_se
covered = (lo <= true_call <= hi)
print(f"MC={mc_price:.6f}, SE={mc_se:.6f}, 95% CI=({lo:.6f}, {hi:.6f}), "
      f"analytic={true_call:.6f}, covered={covered}")
```

## Convergence ($O(1/\sqrt(N))$)
We expect:
\begin{equation}
|\hat{C}_N - C| = \mathcal{O}(N^{-1/2}).
\end{equation}

```python
Ns = np.unique(np.round(np.logspace(3, 6, 12)).astype(int))
est, se, err = [], [], []

# fresh RNG for fair scaling across N
base_rng = np.random.default_rng(123)

for n in Ns:
    # use independent streams (split by SeedSequence)
    child_rng = np.random.default_rng(base_rng.integers(0, 2**32 - 1))
    p, s = mc_euro_call(S0, K, r, sigma, T, n=n, rng=child_rng)
    est.append(p)
    se.append(s)
    err.append(abs(p - true_call))

est, se, err = map(np.asarray, (est, se, err))
```

```python
# %%
plt.figure(figsize=(7,5))
plt.xscale("log")
plt.plot(Ns, est, marker="o", label="MC")
plt.hlines(true_call, xmin=Ns.min(), xmax=Ns.max(), linestyles="dashed", label="Analytic")
plt.fill_between(Ns, est-1.96*se, est+1.96*se, alpha=0.2, label="95% CI")
plt.xlabel("Number of paths N (log)")
plt.ylabel("Call price")
plt.xlim(Ns.min()/2, Ns.max()*2)
plt.title("MC price vs Analytic (95% CI)")
plt.legend()
plt.tight_layout()
plt.show()

# %%
plt.figure(figsize=(7,5))
plt.xscale("log"); plt.yscale("log")
plt.plot(Ns, err, marker="o", label="|MC - Analytic|")
# 1/sqrt(N) guide (scaled to first point)
guide = err[0] * math.sqrt(Ns[0]) / np.sqrt(Ns)
plt.plot(Ns, guide, linestyle="dotted", label="∝ 1/√N (guide)")
plt.xlabel("N (log)"); plt.ylabel("Absolute error (log)")
plt.title("Convergence rate")
plt.legend()
plt.tight_layout()
plt.show()
```
